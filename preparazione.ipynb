```python
import pandas as pd #importo pandas
from sklearn.model_selection import train_test_split #mi serve per fare lo split
```

```python
df= pd.read_csv('heart_failure_clinical_records_dataset.train.csv') #leggo il file csv
```

```python
df.head(10) #faccio una stampa di prova per vedere il file
```

```python
df.info() #vedo le caratteristiche dei dati
```

```python
# Controllo la presenza di valori nulli
print(df.isnull().sum(axis = 0))
```

```python
train, test = train_test_split(df) #faccio lo split dei dati
```

```python
# Controllo la presenza di valori nulli
print(test.isnull().sum(axis = 0))
```
